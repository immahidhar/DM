```python
import pandas
import matplotlib.pyplot as plt 
```

# RFM output

```python
rfmOutput = pandas.read_csv("RFM_output.csv")
rfmOutput
```

```python
rfmOutput.Segment.value_counts()
```

```python
segment = rfmOutput[rfmOutput['Segment'] == 'Churned Customers']
unitPrice = segment['UnitPrice']
recency = segment['Recency']
frequency = segment['Frequency']
daysBtwVisit = segment['DaysBtwVisit']
data = [recency, frequency, unitPrice, daysBtwVisit]
fig = plt.figure(figsize=(10,2))
ax = fig.add_axes([0, 0, 1, 1])
bp = ax.boxplot(data, vert=0)
ax.set_yticklabels(['Recency', 'Frequency', 'Unit Price', 'DaysBtwVisit'])
plt.xlim(-5, 400)
plt.title('RFM - Churned Customers box plot') 
plt.show()
```
